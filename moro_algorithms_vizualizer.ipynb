```python
import math

def CDF(x):
    k = 1/(1+0.2316419*x)
    k_sum = k*(0.319381530 + k*(-0.356563782 + k*(1.781477937 + k*(-1.821255978 + 1.330274429*k))))

    if x >= 0:
        return 1 - (1/math.sqrt(2*math.pi))*math.exp(-(x**2)/2) * k_sum
    else:
        return 1 - CDF(-x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 100)
Y = []
for x in X:
    Y.append(CDF(x))

plt.plot(X, Y)
```

```python
import scipy.stats as st

plt.plot(X, st.norm.cdf(X))
```
